```python
df = df.dropna(subset=['CustomerID'])
```

```python
df = df[~df['Description'].str.contains('test', case=False, na=False)]
df.head(10)
```

```python
df = df[(df['Quantity'] > -100) & (df['Quantity'] < 100)]
df
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# 拆分日期维度
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['OrderDate'] = df['InvoiceDate'].dt.date
df['Hour'] = df['InvoiceDate'].dt.hour
df['Weekday'] = df['InvoiceDate'].dt.weekday 
df
df.to_csv('./cleaned_orders.csv', 
          index=False,       # 不保存行索引
          encoding='utf-8-sig')  # 保证中文不乱码

```

```python
# 创建24小时x7天的数据透视表
hour_weekday = df.pivot_table(
    index='Hour', 
    columns='Weekday', 
    values='InvoiceNo', 
    aggfunc='count',
    fill_value=0
)

# 绘制热力图
plt.figure(figsize=(12,6))
sns.heatmap(
    hour_weekday, 
    cmap='YlGnBu',
    linewidths=0.5,
    annot=True,  # 显示数值
    fmt='d'      # 整数格式
)
plt.title('订单时段分布热力图（周一至周日）')
plt.xlabel('星期（0=周一, 6=周日）')
plt.ylabel('小时')
plt.savefig('./hourly_orders_heatmap.png')
```

```python
# 计算RFM指标
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': 'max',   # 最近消费时间
    'InvoiceNo': 'nunique', # 消费次数
    'TotalPrice': 'sum'     # 消费总额
}).reset_index()

# 重命名列
rfm.columns = ['CustomerID', 'LastPurchase', 'Frequency', 'Monetary']

# 计算R值（距今天数）
rfm['R'] = (pd.to_datetime('now') - rfm['LastPurchase']).dt.days

# 数据分箱（5分制）
def safe_qcut(series, q, labels):
    try:
        return pd.qcut(series, q=q, labels=labels, duplicates='drop')
    except ValueError:
        # 当唯一值不足时自动降级为等宽分箱
        return pd.cut(series, bins=q, labels=labels)

# 应用安全分箱方法
rfm['R_Score'] = safe_qcut(rfm['R'], q=5, labels=[5,4,3,2,1])
rfm['F_Score'] = safe_qcut(rfm['Frequency'], q=5, labels=[1,2,3,4,5]) 
rfm['M_Score'] = safe_qcut(rfm['Monetary'], q=5, labels=[1,2,3,4,5])

# 处理可能出现的NaN（当分箱数超过唯一值数量时）
rfm['F_Score'] = rfm['F_Score'].cat.add_categories(0).fillna(0).astype(int)

# 生成RFM总分
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

# 用户分层规则
def rfm_segment(row):
    if row['R_Score'] >= 4 and row['F_Score'] >=4 and row['M_Score'] >=4:
        return '高价值用户'
    elif row['R_Score'] >=3 and (row['F_Score'] + row['M_Score']) >=6:
        return '潜力用户' 
    else:
        return '一般用户'

rfm['Segment'] = rfm.apply(rfm_segment, axis=1)
# DataFrame变量名为rfm
rfm.to_csv('./rfm_segments.csv',
           index=False,      # 不保存用户ID外的其他索引
           columns=['CustomerID', 'R', 'Frequency', 'Monetary', 'Segment'],  # 选择需要导出的列
           encoding='utf-8-sig')


rfm
```

```python
# 按商品统计销售额
product_sales = df.groupby('StockCode')['TotalPrice'].sum().sort_values(ascending=False)

# 计算累计占比
product_sales_cumsum = product_sales.cumsum()
total_sales = product_sales.sum()
pareto = (product_sales_cumsum / total_sales) * 100

# 绘制帕累托图
fig, ax1 = plt.subplots(figsize=(10,6))
ax1.bar(range(len(product_sales)), product_sales)
ax1.set_ylabel('销售额', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
ax2.plot(range(len(pareto)), pareto, 'r-')
ax2.set_ylabel('累计占比 (%)', color='r')
ax2.tick_params('y', colors='r')

plt.title('商品销售额帕累托分析')
plt.savefig('./pareto_analysis.png')
```
